```
import numpy as np
import matplotlib.pyplot as plt

while True:
    try:
        # Solicitud de datos
        Cuerda = float(input("Digite la longitud de la cuerda (m): "))
        Altura = float(input("Digite la altura del punto de suspensi√≥n (m): "))
        angulos_str = input("Digite los √°ngulos iniciales en grados (separados por comas): ")
        angulos = [float(a.strip()) for a in angulos_str.split(",")]

        g = 9.81  # Aceleraci√≥n de la gravedad (m/s¬≤)

        alcances_max = []  # Para guardar el alcance m√°ximo de cada √°ngulo

        for Angulo in angulos:
            theta0 = np.deg2rad(Angulo)
            theta = np.linspace(0, theta0, 200)

            # Ecuaci√≥n del alcance R(Œ∏)
            R = (2 * Cuerda * (np.cos(theta) - np.cos(theta0)) * np.cos(theta) *
                (np.sin(theta) + np.sqrt(np.sin(theta)**2 +
                (Altura + Cuerda * (1 - np.cos(theta))) / (Cuerda * (np.cos(theta) - np.cos(theta0))))) +
                Cuerda * np.sin(theta))

            # Guardamos el valor m√°ximo del alcance para ese √°ngulo
            R_max = np.nanmax(R)  # por si hay valores no v√°lidos
            alcances_max.append(R_max)

            # Calcular Œ∏_m
            a = (Altura + Cuerda*(1 - 2*np.cos(theta0))) / Cuerda
            b = 0
            c = -(Altura + Cuerda*(1 - np.cos(theta0))) / Cuerda

            coef = [1, a, b, c]
            raices = np.roots(coef)
            x_real = np.real(raices[np.isreal(raices)])[0]
            theta_m = np.arccos(x_real)

            print(f"√Ångulo inicial {Angulo}¬∞ ‚Üí Œ∏_m = {np.degrees(theta_m):.2f}¬∞ | Alcance m√°ximo ‚âà {R_max:.2f} m")

        # üîπ Gr√°fico de barras
        plt.figure(figsize=(8, 5))
        plt.bar([str(a) + "¬∞" for a in angulos], alcances_max, color='skyblue', edgecolor='black')
        plt.xlabel('√Ångulo inicial (¬∞)')
        plt.ylabel('Alcance m√°ximo (m)')
        plt.title('Alcance m√°ximo del salto de Tarz√°n para distintos √°ngulos')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

    except ValueError:
        print("Entrada inv√°lida. Por favor, digite n√∫meros v√°lidos.")

```
